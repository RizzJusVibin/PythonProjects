```python
import math

# Read the instructions to see what to do!

alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

# PART 1
# These functions are provided for you!
def mod_inverse_helper(a, b):
    q, r = a//b, a%b
    if r == 1:
        return (1, -1 * q)
    u, v = mod_inverse_helper(b, r)
    return (v, -1 * q * v + u)

def mod_inverse(a, m):
    assert math.gcd(a, m) == 1, "You're trying to invert " + str(a) + " in mod " + str(m) + " and that doesn't work!"
    return mod_inverse_helper(m, a)[1] % m


# These are the functions you'll need to write:
'''
First line creates a function with the constraints of variable text, a, and b
Second line makes an empty list
Third line makes a for loop, for the index in the range of the length of varible text:
Fourth Line creates a variable, and in that variable it will find the index of text in the index of alpha
Fifth line creates a variable where the found index is multiplied by variable a equal to 3
Sixth line takes the previous number and adds b equal to 9, and the %26 makes sure that it stays within 26 numbers.
Seventh line takes that letter and appends it to the enc list to create a word
Lastly the function will return that list of letters.
'''
def affine_encode(text, a, b):
  enc_list=[]
  for i in range(len(text)):
    m_num = alpha.index(text[i])
    k_num = (m_num * a) 
    e_num = (k_num + b) %26
    enc_list.append(alpha[e_num])
  return ''.join(enc_list)

'''
First line creates a function with the constraints of variable text, a, and b
Second line makes an empty list
Third line makes a for loop, for the index in the range of the length of varible text:
Fourth Line creates a variable, and in that variable it will find the index of text in the index of alpha
Fifth line creates a variable where variabe b = 9 is subtracted from the index.
Sixth line takes the previous number and multiplies it by the inverse of variable a, and the %26 makes sure that it stays within 26 numbers.
Seventh line takes that letter and appends it to the dec list to create a word
Lastly the function will return that list of letters.
'''
def affine_decode(text, a, b):
  dec_list=[]
  for i in range(len(text)):
    m_num = alpha.index(text[i])
    k_num = (m_num-b) 
    e_num = (k_num*mod_inverse(a,26)) %26
    dec_list.append(alpha[e_num])
  return ''.join(dec_list)
'''
These are the variables and the call functions, and that would be the end of this part of the affine cipher.
The first line creates test to say HELLOWORLD
The second line creates a variable a = 3
The third line creates a variable b = 10
The fourth line creates enc = function affine_encode
The fifth line creates dec = function affince_decode
The sixth line prints variable enc
The seventh line prints variable dec
'''
test = "HELLOWORLD"
a = 3
b = 10
enc = affine_encode(test, a, b)
dec = affine_decode(enc, a, b)
print(enc)
print(dec)
# If this worked, dec should be the same as test!



# PART 2
# These  are the functions you'll need to write:
'''
The first line creates a function with the constraint of n_gram
The second line creates a for loop, for the index in the range of the length of test:
Third line creates a function, it finds the index of test in the index of alpha
Fourth line creates a new variable that multiplies 26 by the value of i which goes up by 1 every time the loop repeats
The fifth line creates a new variable that multiplies the previous two variables together.
The sixth line creates a new variable to update the previous variable everytime the loop repeats
The seventh line returns the updated number.
'''
def convert_to_num(ngram):
  f_num=0
  for i in range(len(ngram)):
    k = alpha.index(ngram[i])
    exp = 26**i
    f_num = f_num + k * exp #a =+ b -> a=a+b
  return f_num
'''
The first line creates a function that has the constraints of variables num and n
The second line creates a list
The third line creates a for loop, for the index in the range of n
The fourth line creates f_let to be in num mod 26
The fifth line has num be in mod 26
The sixth line has the letter in text list append to the new list f_let
The seventh line joins the text_list together to form a word
'''
def convert_to_text(num, n):
  text_list=[]
  for i in range(n):
    f_let = num%26
    num = (num//26)
    text_list.append(alpha[f_let])
  return''.join(text_list)

'''
The first line creates the test word to be THEQUICKBROWNFOXJUMPEDOVERTHELAZYDOG
The second line equals the length of test
The third line equals function num to equal convert_to_num(test)
The fourth line equals function answer to equal convert_to_text(num,l)
The fifth line prints variable num
The sixth line prints variable answer
'''
test = "THEQUICKBROWNFOXJUMPEDOVERTHELAZYDOG"
l = len(test)
num = convert_to_num(test)
answer = convert_to_text(num, l)
print(num)
print(answer)
# If this worked, answer should be the same as test!



# PART 3

# These are the functions you'll need to write:
def affine_n_encode(text, n, a, b):#Creates a new function
  r = len(text)%n #how many grams can we split the text in
  more = n - r #r takes the remainder of %26, just so we know how many letters to add
  final_text = text+"X"*more#Creates final text to be text + X times more
  encod_list = []#Creates a new list
  for location in range(0,len(text),n):#The three points in the range go in this order, Start, End, and increment.
    ngram = final_text[location:location+n]#Will split the text into letters of 5
    num = convert_to_num(ngram)#Calls the function to convert the group of letters to a number
    af_num=(num*a+b)%(26**n)#This takes the previous number and usees the affine cipher multiply and add
    af_let = convert_to_text(af_num,n)#Converts the affine number to a text, so the encrypted letters
    encod_list.append(af_let)#We make the long list of letters that we get with the groups of 5.
  return ''.join(encod_list)#This will return the enoded list 

def affine_n_decode(text, n, a, b):#Creates a new variable
  dec_list = []#Creates a new list
  a_inv = mod_inverse(a,26**n)#Variable a_inv = mod_inverse(a,26**n) 
  for i in range(0,len(text),n):#Creates a for loop for i in range of 0, length of text, n
    ngram = text[i:i+n]#Variable ngram = text index : index + n 
    num=convert_to_num(ngram)#num = function convert_to_num
    af_num = (num-b)*a_inv%(26**n)#af_num = variable num - b times a_inv mod 26 to the power of n
    af_le = convert_to_text(af_num,n)#af_le = function convert_to_text
    dec_list.append(af_le)#Will append the letters to the list
  return ''.join(dec_list)#Will join the list together

test = "THEQUICKBROWNFOXJUMPEDOVERTHELAZYDOG"#Variable test = THEQUICKBROWNFOXJUMPEDOVERTHELAZYDOG
n = 5 #n=5
a = 347 #a=347
b = 1721 #b=1721
enc = affine_n_encode(test, n, a, b) #enc = function affince_n_encode
dec = affine_n_decode(enc, n, a, b) #dec = function affince_n_decode
print(enc) #Will print enc
print(dec) #Will print dec
# If this worked, dec should be the same as test!
```
