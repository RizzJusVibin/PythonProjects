#Caesar Shift

This python program uses the skills of using lists, functions, text manipulation in order to write out a Caesar Shift that helps encode messages.

```python
# Read the instructions to see what you need to do here!

alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

def caesar_encode(text, n):
  answer_list = []
  for letter in text: #letter is a varible and is counting in text, example what is 0 index in text, = t
    ind = alpha.index(letter)# What is that index of the letter in alpha
    ind +=n #here we are adding the shift to the index.
    ind %=26# ind = ind%26
    answer_list.append(alpha[ind])
  return "".join(answer_list)#return the strings fromt he list


def caesar_decode(text, n):
  answer_list = []
  for letter in text:
    ind = alpha.index(letter)
    ind -=n # or ind == n
    ind %=26
    answer_list.append(alpha[ind])
  return "".join(answer_list)


text = input("Enter text:")
text = text.upper()


test = "HELLOWORLD"
shift = 1
enc = caesar_encode(text, shift)
dec = caesar_decode(enc, shift)
print(enc)
print(dec)
 #If this worked, dec should be the same as test!
```
